```python
def add(v1, v2):
    return [v1[0] + v2[0], v1[1] + v2[1]]
def length(v):
    return sqrt(v[0]**2 + v[1]**2)    

dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]
dino_vectors2 = [add(v, (-1.5, -2.5)) for v in dino_vectors]

# Points.__init__ 接收可变参数 *vectors，* 会将 dino_vectors 列表的元素逐个传递；
draw(Points(*dino_vectors))
```

```python
draw(
    Points(*[(x, x**2) for x in range(-10, 11)]),
    grid=(1, 10), # draws vertical grid lines every one unit and horizontal grid lines every ten units
    nice_aspect_ratio=False # don't require x scale to match y scale
)
```

```python
draw(
    Points(*dino_vectors),
    Segment((6, 4), (3, 1)),
)
```

```python
arrows = [Arrow(tail, tip, color=black) for tip, tail in zip(dino_vectors, dino_vectors2)]
draw(
    Points(*dino_vectors, color=blue),
    # 按列表中的顺序两两连接相邻的点
    Polygon(*dino_vectors, color=blue),

    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red),
    *arrows,
)
```

```python
p1 = 4, 3
p2 = -1, 1
p1_p2_sum = tuple(map(lambda x, y: x + y, p1, p2))
print(p1)
draw(
    Arrow(p1, color=purple),
    Arrow(p2, color=red),
    Arrow(p1_p2_sum, color=black),
    Arrow(p1_p2_sum, p1, color=red),
    Points(p1, p2, p1_p2_sum, color=blue, show_coordinates=True),
    super_title="Tip-to-tail addition: {0}+{1}".format(p1, p2),
)
```

```python
p1 = 3, 1
p2 = 1, 2
draw(
    Points(p1, p2, color=blue, show_coordinates=True),
    Arrow(p1, color=black),
    axes=False,
    grid=None,
    annotate_origin=True,
)
```

```python
v = 6, 4
v15 = tuple(map(lambda x: x * 1.5, v))
draw(
    Arrow(v, color=purple),
    Points(v, color=blue, annotation="v"),
    Arrow(v15, color=purple),
    Points(v15, color=blue, annotation="1.5 • v"),

    Arrow((v[0], 0), color=red),
    Points((v[0], 0), color=blue),
    Arrow(v, (v[0], 0), color=blue),
    Points((v15[0], 0), color=blue),
    Arrow((v15[0], 0), color=red),
    Arrow(v15, (v15[0], 0), color=blue),
)
```

```python
v1 = -2, 1
negatedV1 = tuple(map(lambda x: -x, v1))
draw(
    Points(v1, color=blue, annotation="v", show_coordinates=True),
    Points(negatedV1, color=blue, annotation="-v", show_coordinates=True),

    Arrow(v1, color=red),
    Arrow(negatedV1, color=blue),
)
```

```python
p1 = -1, 3
p2 = 2, 2
negatedP2 = tuple(map(lambda x: -x, p2))
p1_p2_sum = tuple(map(lambda x, y: x - y, p1, p2))

draw(
    Arrow(p1, color=blue),
    Arrow(p2, color=blue),
    Arrow(p1_p2_sum, color=red),
    Arrow(p1, p2, color=red),
    Points(p1, p2, p1_p2_sum, color=blue, show_coordinates=True),
    Points(p1, annotation="v", color=blue),
    Points(p2, annotation="w", color=blue),
    Points(p1_p2_sum, annotation="v - w", color=blue),
    Points(negatedP2, annotation="-w", color=blue),
    Arrow(negatedP2, color=purple),
    Arrow(p1_p2_sum, p1, color=purple),
    super_title="v-w, the displacement from w to v",
)
```
